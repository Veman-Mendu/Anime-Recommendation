```python
import pandas as pd

data = pd.read_csv('S:\Projects/anime\data/up_data.csv')
data.head(10)
```

```python
c = data['Scores'].mean()
print(c)
```

```python
def weight_rating(x,c=c):
    f = x['Favorites']
    f = int(f)
    m = x['Members']
    m = int(m)
    s = x['Scores']
    s = int(s)
    r = (f/m)*(s/c)
    return r
```

```python
def weight_rating_2(x,c=c):
    f = x['Favorites']
    f = int(f)
    m = x['Members']
    m = int(m)
    s = x['Scores']
    s = int(s)

    return (f/m)*(c/s)
```

```python
data['rec_score'] = data.apply(weight_rating,axis=1)
data.head(10)
```

```python
sort_data = data.sort_values('rec_score',ascending=False)
sort_data.head(20)
```
