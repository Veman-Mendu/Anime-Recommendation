```python
genre = input("Please select a genre")
studio = input("Please select a studio")
```

```python
import pandas as pd
data = pd.read_csv('S:\Projects/anime\data/dup_data.csv')
dup_data = data.copy()

animes = dup_data[(dup_data['Genre'] == genre) | (dup_data['Studio'] == studio)]
```

```python
c = animes['Scores'].mean()
```

```python
def weight_rating(x,c=c):
    f = x['Favorites']
    f = int(f)
    m = x['Members']
    m = int(m)
    s = x['Scores']
    s = int(s)
    r = (f/m)*(s/c)
    return r
```

```python
animes['rec_score'] = animes.apply(weight_rating,axis=1)
sort_data = animes.sort_values('rec_score',ascending=False)
sort_data
```
