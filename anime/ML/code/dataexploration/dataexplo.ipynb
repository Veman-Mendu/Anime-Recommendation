```python
import pandas as pd

animes = pd.read_csv("S:\Projects/anime\data\data_1_500.csv")
animes.head(10)
```

```python
#Below is the code to find the number of null values and data type of all features.
animes.info()
```

```python
#Since we have given NA for all the null values in the data. Let's try finding no.of NA's in each column
columns = ['Episodes', 'Status', 'Studios', 'Genres', 'Ratings', 'Members', 'Favorites', 'titles']

for col in columns:
    count = 0
    if(animes[col].str.contains('NA').any() == True):
        count = count + 1
    print(f'{col} : {count}')
```

```python
#Below is the details about the numeric features present in the dataset
animes.describe()
```

```python
#In the info it shows that members and favorites are object data types but they need to be in int or float formats.
#animes['Members'] = animes['Members'].astype(int,errors = 'raise') -- This doesnot work since the numbers have , in between them.
members = []

for member in animes['Members']:
    member = member.replace(',','')
    members.append(member)
    
animes['Members'] = pd.Series(members).astype(int, errors = 'raise')
animes.info()
```

```python
#In the info it shows that members and favorites are object data types but they need to be in int or float formats.
favorites = []

for favorite in animes['Favorites']:
    favorite = favorite.replace(',','')
    favorites.append(favorite)
    
animes['Favorites'] = pd.Series(favorites).astype(int, errors = 'raise')
animes.info()
```
