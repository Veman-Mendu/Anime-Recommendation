```python
import pandas as pd
data = pd.read_csv('S:\Projects/anime\data/up_data.csv')
```

```python
genres = []
for i in range(300):
    genre = data.iloc[i]["Genres"].replace(' ','')
    genre = genre.split(",")
    ge = []
    for g in genre:
        g = g.lower()
        ge.append(g)
    genres.append(ge)
```

```python
studios = []
for i in range(300):
    studio = data.iloc[i]["Studios"].replace(' ','')
    studio = studio.split(",")
    st = []
    for s in studio:
        s = s.lower()
        st.append(s)
    studios.append(st)
```

```python
import ast

sp = data['SP']
spn = []
for i in range(300):
    s = data.iloc[i]["SP"]
    s = str(s)
    if s == 'nan':
        s = []
    else:
        s = ast.literal_eval(s)
    spl = []
    for i in s:
        i = i.lower()
        spl.append(i)
    spn.append(spl)
```

```python
dup_data = data
dup_data = dup_data.drop(columns=['Genres','Studios'])
dup_data['Genre'] = genres
dup_data['Studio'] = studios
dup_data['SP'] = spn
dup_data
```
