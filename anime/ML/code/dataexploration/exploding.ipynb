```python
import pandas as pd
data = pd.read_csv('S:\Projects/anime\data/up_data.csv')
```

```python
#The following code is for exploding Genres field.
genres = []
for i in range(300):
    genre = data.iloc[i]["Genres"].replace(' ','')
    genre = genre.split(",")
    genres.append(genre)
```

```python
#The following code is for exploding studios field
studios = []
for i in range(300):
    studio = data.iloc[i]["Studios"].replace(' ','')
    studio = studio.split(",")
    studios.append(studio)
```

```python
dup_data = data
dup_data = dup_data.drop(columns=['Genres','Studios'])
dup_data['Genre'] = genres
dup_data['Studio'] = studios
dup_data = dup_data.explode('Genre')
dup_data = dup_data.explode('Studio')
dup_data
```

```python
dup_data.to_csv('S:\Projects/anime\data/dup_data.csv',index=False)
```
